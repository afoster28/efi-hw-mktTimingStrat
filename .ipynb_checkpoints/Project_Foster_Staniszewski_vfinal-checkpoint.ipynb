We will, however, **create a new feature** for our machine learning solution based on which it could automatically adjust to these small series.

```python
grouped_frame = grouped.to_frame()
grouped_frame.rename(columns = {'Close':'TimepointsCount'}, inplace=True)
grouped_frame = grouped_frame.assign(FewTimepoints = lambda x : x.TimepointsCount < 1201)
grouped_frame
```

```python
stock_prices = stock_prices.merge(grouped_frame, on='SecuritiesCode', how='left')
stock_prices
```

Let's add features based on moving averages. This should be useful when determining stock performance/trends

```python
concat_parts = []
for seccode in list(grouped.keys()): # iterate over all security codes 
    # create a view (reference) to rows by a given security code. 
    this_seccode_prices = stock_prices.loc[stock_prices.SecuritiesCode == seccode].copy()
    # will create SMA assuming sorted by date, so sort just in case
    this_seccode_prices.sort_values(by=['Date'])
    # modify dataframe view (related to currently processed seccode) by adding SMA columns
    this_seccode_prices["SMA 15"] = this_seccode_prices[["Close"]].rolling(15).mean().shift(1)
    this_seccode_prices["SMA 60"] = this_seccode_prices[["Close"]].rolling(60).mean().shift(1)
    concat_parts.append(this_seccode_prices)

stock_prices = pd.concat(concat_parts).reset_index(drop=True)
stock_prices
```

All Security Codes are processed now and have SMA 15 and SMA 60 columns. We can see it in action for a random security code.

```python
# random_security_code = 1333
random_security_code = np.random.choice(stock_prices['SecuritiesCode'])
stock_prices[stock_prices['SecuritiesCode'] == random_security_code].filter(['Close', 'SMA 15', 'SMA 60']).plot().grid()
```

Let's do the same for returns and log-returns. We're adding log-returns as a feature as this is commonly used when modelling price moves. Also, it's a non-linear function of the feature, so regression will not pick it up automatically and a feature like that will introduce new information (it's not a linear combination of other features).

```python
concat_parts = [] #shadowing on purpose
for seccode in list(grouped.keys()): # iterate over all security codes 
    # create a view (reference) to rows by a given security code. 
    this_seccode_prices = stock_prices.loc[stock_prices.SecuritiesCode == seccode].copy()
    # will create SMA assuming sorted by date, so sort just in case
    this_seccode_prices.sort_values(by=['Date'])
    # returns as the pct change of day-on-day moves
    this_seccode_prices["Return"] = this_seccode_prices["Close"].pct_change(1)
    # handle log of zero and negative numbers in a special way - assign zero.
    this_seccode_prices["LogReturn"] = 0
    this_seccode_prices["LogReturn"] = np.log(this_seccode_prices["Return"], where=this_seccode_prices["Return"] > 0)
    concat_parts.append(this_seccode_prices)

stock_prices = pd.concat(concat_parts).reset_index(drop=True)
stock_prices
```

```python
target_check = stock_prices[['Target','Return']]
corr_matrix2 = target_check.corr(numeric_only=True).round(2)
sns.set(rc={"figure.figsize":(16, 6)})
sns.heatmap(corr_matrix2, annot=True)
plt.show()
```

We have observed that the Target variable in the dataset has nothing to do with Return calculated on DoD close. We will therefore remove Target and keep Return as out dependent variable, based on expert judgment and financial knowledge.

Now let's add volatility of returns.

```python
concat_parts = [] #shadowing on purpose
for seccode in list(grouped.keys()): # iterate over all security codes 
    # create a view (reference) to rows by a given security code. 
    this_seccode_prices = stock_prices.loc[stock_prices.SecuritiesCode == seccode].copy()
    # will create MSD assuming sorted by date, so sort just in case
    this_seccode_prices.sort_values(by=['Date'])
    this_seccode_prices["MSD 15"] = this_seccode_prices[["Return"]].rolling(15).std().shift(1)
    this_seccode_prices["MSD 60"] = this_seccode_prices[["Return"]].rolling(60).std().shift(1)
    concat_parts.append(this_seccode_prices)

stock_prices = pd.concat(concat_parts).reset_index(drop=True)
stock_prices
```

```python
corr_matrix = stock_prices.corr(numeric_only=True).round(2)
sns.set(rc={"figure.figsize":(16, 6)})
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

We see that Open/High/Low/Close - so prices - are heavily correlated with SMAs. This is expected, as SMAs are constructed from those prices. Expected Dividend column is correlated with prices too, and with SMAs consequently. We observe large negative correlation between TimepointsCount and FewTimepoints, but this is also expected as one feature is directly a function of another. # 3 - Handling NaN Values

Let's inspect columns which have lots of NaNs instead of proper values. It's time for inputation, so we need to explore these more. 

```python
nans_per_column = stock_prices.isna().sum(axis = 0).divide(len(stock_prices))
nans_per_column.loc[nans_per_column > 0.5]
```

The above columns have more NaNs than meaningful values. It is, however, possible that for rows that contain non-NaN entries in columns above, this information has high predictive power. We will explore the statistical significance of parameters with special consideration for the columns above.

Let's replace NaNs with numeric values so that we can use these columns in ML models that we'll soon implement.

```python
nans_per_column.loc[nans_per_column > 0.0]
```

Dividends can easily be replaced with zeroes. The same goes for SMAs and MSDs (since NaN values are inserted at the beginning of the time series data when there's not enough elements in the window to calculate rolling averages). Extra columns coming from financials can also be zeroed by intuitive understanding of the meaning of these features, i.e. No earnings per share (this is detrimental!), no changes in subsidiaries and so on.

We are aware that zeroing out nulls can potentially affect results, however with this volume of nulls we wouldn't have any viable replacement value for most of these cases as there simply isn't enough accurate data to use, e.g. for linear interpolation or mean.

Even for variables where very few observations are null, those would have time series starting later than the starting point, as discovered earlier.

Let's take a look at returns that are NaNs.

```python
stock_prices[stock_prices["Return"].isnull()].reset_index(drop=True)
```

These must be elements at the beginning of the time series (no SMAs, so very early time points). Return is a NaN, because there is no earlier value to get the return from - these entries are very first time series elements. Let's drop them and re-evaluate NaNs in the dataframe.

```python
stock_prices_dropped_na_returns = stock_prices[stock_prices['Return'].isna() == False]
stock_prices = stock_prices_dropped_na_returns
nans_per_column = stock_prices.isna().sum(axis = 0).divide(len(stock_prices))
nans_per_column.loc[nans_per_column > 0.0]
```

Let's impute 0 everywhere now, as discussed before.

```python
stock_prices = stock_prices.fillna(0)
nans_per_column = stock_prices.isna().sum(axis = 0).divide(len(stock_prices))
nans_per_column
```

# 4 - Normalisation

Define variables

```python
dependent_variable_str = 'Return'
dependent_variable = [dependent_variable_str]
nonnumeric_variables = ['ExpectedDividend', 'ForecastEarningsPerShare', 'MaterialChangesInSubsidiaries', 'NumberOfIssuedAndOutstandingSharesAtTheEndOfFiscalYearIncludingTreasuryStock', 'NumberOfTreasuryStockAtTheEndOfFiscalYear']
explanatory_variables = ['SMA 15', 'SMA 60', 'MSD 15', 'MSD 60', 'High', 'Low'] + nonnumeric_variables
explanatory_variables_norm = [var + ' ZScore' for var in explanatory_variables]
variables = dependent_variable + explanatory_variables + explanatory_variables_norm
```

Replace -- with zeroes and convert financial non-numeric columns to floats

```python
stock_prices.replace(to_replace='', value=0, inplace=True)
for col in nonnumeric_variables:
    stock_prices[col] = stock_prices[col].astype(float)
```

As some of the explanatory variables have a large domain (e.g. prices, volumes), standardisation is required to improve their predictive power. This will be particularly useful for KNN.

```python
concat_parts = [] #shadowing on purpose
for seccode in list(grouped.keys()): # iterate over all security codes 
    # create a view (reference) to rows by a given security code. 
    this_seccode_prices = stock_prices.loc[stock_prices.SecuritiesCode == seccode].copy()
    for col in explanatory_variables:
        col_mean = this_seccode_prices[col].mean()
        col_std = this_seccode_prices[col].std()
        this_seccode_prices[col + ' ZScore'] = (this_seccode_prices[col] - col_mean) / col_std
    concat_parts.append(this_seccode_prices)

stock_prices = pd.concat(concat_parts).reset_index(drop=True)
stock_prices = stock_prices.fillna(0)
stock_prices
```

```python
normalised_variable = stock_prices['ExpectedDividend ZScore']

bins = np.arange(-10, 10, 0.01)

# plt.xlim([min(normalised_variable) - 0.1, max(normalised_variable) + 0.1])
plt.xlim([-0.5, 0.5])

plt.hist(normalised_variable, bins=bins, alpha=0.5)
plt.title('Normalised Variable Distribution')
plt.xlabel('ExpectedDividend_ZScore (bin size = 0.1)')
plt.ylabel('Count')

plt.show()
```

```python
min(normalised_variable), max(normalised_variable)
```

Roughly shaped like a standard normal distribution - slightly tilted to the left of 0 due to large outlier at the end of the distribution

```python
normalised_variables = [col for col in stock_prices.columns if 'ZScore' in col]
for col in normalised_variables:
    print(min(stock_prices[col]), max(stock_prices[col]))
```

# 5 - Statistical Significance Tests

It is worth running OLS on all features to check each feature's significance prior to running through the final models.

```python
x = stock_prices.drop(['Return', 'Target', 'Date'], axis=1)
y = stock_prices['Return']
# Percentage train set
proportion_in_train = 0.8
split = int(proportion_in_train*len(stock_prices))

# Train set creation
x_train = x.iloc[:split]
y_train = y.iloc[:split]

x_test = x.iloc[split+1:]
y_test = y.iloc[split+1:]
```

```python
x_train = sm.add_constant(x_train)
model = sm.OLS(y_train,x_train.astype(float)).fit()
print(model.summary())
```

```python
coeffs = model.params.drop(labels=['const'])
cols = x_train.columns.drop(labels=['const'])

importances = pd.DataFrame(data={
    'Attribute': cols,
#     'Importance': model.coef_,
    'Importance': coeffs,
})
importances = importances.sort_values(by='Importance', ascending=False)
plt.bar(x=importances['Attribute'], height=importances['Importance'], color='#087E8B')
plt.title('Feature importance obtained from coefficients', size=20)
plt.xticks(rotation='vertical')
plt.show()
```

Beta parameters are small. Testing for their significance using t-tests will allow us to determine which features can be relevant for the exercise.

```python
# [feature for feature in model.pvalues if abs(feature) < 0.05]
pvalues = pd.DataFrame(data={
    'Attribute': x_train.columns,
    'pvalue': model.pvalues,
})
pvalues = pvalues[abs(pvalues['pvalue']) < 0.05].sort_values(by='pvalue').reset_index(drop=True)
print(pvalues)
```

As noted before, there was some correlation between some of the features and the target - we see the view reinforced following significance testing where each of the above are statistically significant at 5% confidence level. These variables are expected to have non-zero predictive power.

Of the momentum features, an interesting observation is that SMA 15 has a greater expected impact than SMA 60, whilst MSD 60 greater than MSD 15, meaning more recent fluctuations in price yet more distant variability from the mean in price are more likely predictors of returns.

A crucial part of this exercise is applying economic and financial knowledge to further exclude or include features. Without this, the model would be prone to producing spurious results over time and overfitting.

Exclude:
- Close (feature underlying the target)
- LogReturn (transformation of the target)
- RowId (identifier)
- Other statistically insignificant (and indeed economically insignificant) features, e.g. SecuritiesCode, AdjustmentFactor, SupervisionFlag, TimepointsCount

Include:
- Statistically significant features identified above, except for the exclusions noted
- ExpectedDividend (economically significant)
- ForecastEarningsPerShare (economically significant)
- MaterialChangesInSubsidiaries (economically significant)
- NumberOfIssuedAndOutstandingSharesAtTheEndOfFiscalYearIncludingTreasuryStock (economically significant)
- NumberOfTreasuryStockAtTheEndOfFiscalYear (economically significant)

Many of the financial features were identified as statistically insignificant due to the prevalence of zeroes from the lack of data for so many companies, however, we believe that in the unusual occasion when observations are populated they are likely predict a change in return. For example, as EPS is forecast to increase, this could indicate better profitability, a likely increase in stock price and therefore return.

# 6 - Train-Test Split

These are therefore the features used in modelling:

```python
pvalues = pvalues[~pvalues['Attribute'].isin(['const','Close','LogReturn','RowId'])].reset_index(drop=True)
pvalues = pvalues['Attribute'].tolist()
```

```python
pvalues_reg = [col for col in pvalues if 'ZScore' not in col]
explanatory_variables = pvalues_reg + nonnumeric_variables
explanatory_variables
```

```python
pvalues_norm = [col for col in pvalues if 'ZScore' in col]
explanatory_variables_norm = pvalues_norm + [col + ' ZScore' for col in nonnumeric_variables]
explanatory_variables_norm
```

```python
selected_columns = [ 'RowId', 'Date', 'Return', 'SecuritiesCode' ] + explanatory_variables + explanatory_variables_norm

stock_prices_proc = stock_prices[selected_columns]
stock_prices_proc
```

We have a large panel dataset with multiple security returns and their features across mutiple timepoints. We pick one security to perform quick analysis on and to try predicting its return. We note that the analysis can be replicated across as many securities as desired.

```python
stock_prices_sample = stock_prices_proc[stock_prices_proc['SecuritiesCode'] == random_security_code]
stock_prices_sample
```

Now, we select train-test sets. Our rationale for selecting an 80/20 split comes from industry practice. We have not performed subsequent cross validation, as only select types of CV (like rolling CV) are applicable for time series data and its build out would not be expected to yield significantly more accurate or financially intuitive results than the set split. This also ensures the largest possible training set is considered within the 80% portion of the dataset - we view a longer time horizon as generally being helpful in explaining asset moves.

We also split our training sets into regular and normalised versions - the former will be used for linear regression, the latter for KNN and SVR which benefit from smaller differences in domains.

```python
# Percentage train set
proportion_in_train = 0.8
split = int(proportion_in_train*len(stock_prices_sample))

dependent_variable = ['Return']
# Train set
x_train = stock_prices_sample[explanatory_variables].iloc[:split]
x_train_norm = stock_prices_sample[explanatory_variables_norm].iloc[:split]
y_train = stock_prices_sample[dependent_variable].iloc[:split]

# Test set
x_test = stock_prices_sample[explanatory_variables].iloc[split:]
x_test_norm = stock_prices_sample[explanatory_variables_norm].iloc[split:]
y_test = stock_prices_sample[dependent_variable].iloc[split:]
```

# 7 - Linear Regression

```python
# Initialize the class
reg = LinearRegression()

# Fit the model
reg.fit(x_train, y_train)
```

Perform predictions and show entire time horizon

```python
# Create predictions for the whole dataset
x = np.concatenate((x_train, x_test), axis=0)

stock_prices_sample['prediction_reg'] = reg.predict(x)

# Prediction visualisation
plt.plot(stock_prices_sample['Date'], stock_prices_sample['prediction_reg'])
plt.title('OLS Predicted Return')
plt.show()
```

```python
plt.plot(stock_prices_sample['Date'], stock_prices_sample['prediction_reg'], stock_prices_sample['Date'], stock_prices_sample['Return'])
plt.show()
```

Realised returns exhibit more extreme values that their predictions - based on a) slow-changing SMAs, b) difference in scale (SMAs based on Close which is large in absolute terms vs Return which is a relative metric), c) any potential inaccuracy

Compute profit

```python
# Compute the position
stock_prices_sample['position_reg'] = np.sign(stock_prices_sample['prediction_reg'])

# Compute the returns
stock_prices_sample['strategy_reg'] = stock_prices_sample[dependent_variable_str] * stock_prices_sample['position_reg'].shift(1)

(stock_prices_sample['strategy_reg'].iloc[split:].cumsum()*100).plot()
plt.show()
```

The position is created to match the sign of the prediction of returns for a given day. This position is then applied to the following day and the strategy's PnL is determined based on the market moves (return) this following day.

```python
stock_prices_sample.head()
```

```python
plt.plot(stock_prices_sample['Date'], stock_prices_sample['strategy_reg'], stock_prices_sample['Date'], stock_prices_sample['Return'])
plt.show()
```

The graph shows the strategy's daily performance relative to market returns

```python
stock_prices_sample['prediction_svr'] = svr.predict(x_norm)

plt.plot(stock_prices_sample['Date'], stock_prices_sample['prediction_svr'])
plt.title('SVR Predicted Return')
plt.show()
```

# 10 - Conclusion

Now we compare results across models for test data using RMSE and MAE

```python
y_pred_ols = reg.predict(x_test)
```

```python
plt.plot(y_pred_knn, y_test, 'o')
plt.title('KNN vs Actual Return')
plt.xlabel('KNN')
plt.ylabel('Return')
plt.show()
```

```python
rmse_svr = np.sqrt(mean_squared_error(y_test, y_pred_svr))
mae_svr = mean_absolute_error(y_test, y_pred_svr)
model_score_svr = pd.DataFrame([rmse_svr, mae_svr]).T.rename(columns ={0:"RMSE", 1:"MAE"})
model_score_svr.rename(index={0: 'SVR'},inplace=True)
```

```python
plt.plot(y_pred_svr, y_test, 'o')
plt.title('SVR vs Actual Return')
plt.xlabel('SVR')
plt.ylabel('Return')
plt.show()
```
